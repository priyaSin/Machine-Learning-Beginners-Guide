```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
columnNames = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age" , "Y"]
```

```python
diabetes = pd.read_csv("pima-indians-diabetes.csv" , names = columnNames) 
diabetes.head()
```

```python
#You can do the initial preprocessing checks, I m skipping that part for now
# Apply Scaling
scaler = MinMaxScaler()
```

```python
diabetes[diabetes.columns] = scaler.fit_transform(diabetes[diabetes.columns])
diabetes.head()
```

```python
#Adding a new column for b parameter of m1.x1 + m2.x2 .. + b with all feature values to be 1  
diabetes["x"] = [1 for i in range(768)]
diabetes.head()
```

```python
#Rearranging the columns
diabetes = diabetes[["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age" , "x", "Y"]]  
diabetes.head()
```

```python
# Changing the values of Y columns from integer to float
diabetes.Y = diabetes.Y.astype(int)
diabetes.head()
```

```python
#Logistic Function
def logit(x , m):
    z = (x * m).sum()
    return 1/(1 + math.exp(-z))
```

```python
def cost_function(x_train , y_train , m):
    cost = 0
    for i in range(len(x_train)):
        cost += (y_train[i] * math.log(logit(x_train[i] , m)) + (1 - y_train[i]) * math.log(1 - logit(x_train[i] , m)))
    return (-1/len(x_train)) * cost

```

```python
def predict(x_test , m):
    y_predict = np.array([])
    for i in range(len(x_test)):
        y_predict = np.append(y_predict , logit(x_test[i]  , m))
    y_predict = (y_predict >= y_predict.mean()).astype(int)
    return y_predict
```

```python
def cal_score(test , m):
    x_test = test[: , 0:9]
    y_test = test[: , 9]
    y_predict = predict (x_test , m)
    indicator = np.array((y_predict == y_test).astype(int))
    print("Number of samples correctly predicted : " , indicator.sum() , " out of " , len(x_test))  
    score = (indicator.sum()) / len(x_test)
    return score
```

```python
def step_gradient(x_batch , y_batch , m , learning_rate):
    batch_size = len(x_batch)
    for j in range(len(m)):
        s = 0
        for i in range(len(x_batch)):
            s += (logit(x_batch[i] , m) - y_batch[i]) * x_batch[i , j]
        m[j] = m[j] - learning_rate * s
    return m
```

```python
def gradient_step_runner(train, learning_rate , num_of_iterations , starting_m , num_of_batches):  
    m = starting_m
    
    #For creating the plot
    m_0_list = np.array([])
    cost_list = np.array([])
    
    initial_cf = cost_function(train[: , 0:9] , train[: , 9] , m)
    print("Initial Cost : " , initial_cf)
    
    
    #"Training set (having 18 batches with 32 in each batch) shape"
    new_train = np.array(np.array_split(train , num_of_batches))
    x_train = new_train[: , : , 0:9]
    y_train = new_train[: , : , 9]
    
    
    for i in range(num_of_iterations):
        for j in range(len(x_train)):
            m = step_gradient(x_train[j] , y_train[j] , m , learning_rate )
    
        #For creating a plot
        if(i % 10 == 0):
            m_0_list = np.append(m_0_list , m[0])
            cost = cost_function(train[: , 0:9] , train[: , 9] , m)
            cost_list = np.append(cost_list , cost)
    #plot
    plt.scatter(m_0_list , cost_list)
    plt.plot(m_0_list , cost_list)
    plt.show()
        
    
    final_cf = cost_function(train[: , 0:9] , train[: , 9] , m)
    print("Final cost : " , final_cf)

    return m
```

```python
def run(diabetes):
    train , test = train_test_split(diabetes , test_size = 0.25)
    train = train.as_matrix()
    test = test.as_matrix()
    #Training set shape = 576 , 10
    starting_m = np.array([0.0 for i in range(len(train[0]) -1)])
    learning_rate = 0.01
    num_of_iterations = 1000
    num_of_batches = 18
    m = gradient_step_runner(train , learning_rate , num_of_iterations , starting_m, num_of_batches)
    return m
    
```

```python
m = run(diabetes)
print("Final values of m" , m)
score = cal_score(test , m)
print("Score" , score)
```
